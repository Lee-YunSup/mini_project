## 1. 사용자의 충성도 분류 (3그룹) 진행 

```python
df['날짜점수']=0
df['빈도점수']=0
df['가격점수']=0
```

```python
for i in df.index:
    if df.loc[i,'주문일자']<pd.Timestamp('2020-07-01'):
        df.loc[i,'날짜점수']=0
    elif df.loc[i,'주문일자']<pd.Timestamp('2021-01-01'):
        df.loc[i,'날짜점수']=20
    elif df.loc[i,'주문일자']<pd.Timestamp('2021-07-01'):
        df.loc[i,'날짜점수']=40
    elif df.loc[i,'주문일자']<pd.Timestamp('2022-01-01'):
        df.loc[i,'날짜점수']=60
    elif df.loc[i,'주문일자']<pd.Timestamp('2022-07-01'):
        df.loc[i,'날짜점수']=80   
    else:
        df.loc[i,'날짜점수']=100
```

```python
for i in df.index:
    name=df.loc[i,'주문번호']
    m=df['주문번호']==name
    res=df[m].shape[0]
    
    df.loc[i,'빈도점수']=res
```

```python
for i in df.index:
    name=df.loc[i,'주문번호']
    m=df['주문번호']==name
    res=df[m]['판매금액'].sum()
    
    df.loc[i,'가격점수']=res
```

```python
df
```

```python
df_score=df[['날짜점수','빈도점수','가격점수']]
```

```python
df_score
```

```python
df['가격점수'].max()
```

```python
minmax=min_max_scaler.fit_transform(df_score)
```

```python
df_minmax=pd.DataFrame(minmax,columns=['날짜점수min_max','빈도점수min_max','가격점수min_max'])
```

```python
df_minmax
```

```python
df_minmax['합계점수']=df_minmax['날짜점수min_max']+df_minmax['빈도점수min_max']+df_minmax['가격점수min_max']
```

```python
df_minmax.max()
```

```python
df_merge=pd.merge(df,df_minmax,left_index=True, right_index=True)
```

```python
df_drop=df_merge.drop(['날짜점수','빈도점수','가격점수'],axis='columns')
```

```python
df_drop
```

```python
df_final=df_drop.drop_duplicates('주문번호')
```

```python
df_final
```

```python
df_final2=df_final[['주문번호','합계점수']]
```

```python
df_final2
```

```python
q1=df_final2['합계점수'].quantile(1/3)
q2=df_final2['합계점수'].quantile(2/3)
q3=df_final2['합계점수'].quantile(1)
```

```python
print(q1,q2,q3)
```

```python
list1=[]

for i in range(len(df_final2.index)):
    if (df_final2.iloc[i,1]) <= q1:
        list1.append([df_final2.iloc[i,0],df_final2.iloc[i,1],'3등급'])
    elif (df_final2.iloc[i,1]) <=q2:
        list1.append([df_final2.iloc[i,0],df_final2.iloc[i,1],'2등급'])
    else:
        list1.append([df_final2.iloc[i,0],df_final2.iloc[i,1],'1등급'])
final=pd.DataFrame(list1,columns=['주문번호','합계점수','등급'])
final
```

```python
final_sort= final.sort_values('합계점수', ascending=False)
```

```python
f1st=final_sort['등급']=='1등급'
s2nd=final_sort['등급']=='2등급'
t3rd=final_sort['등급']=='3등급'
```
