```python
import matplotlib.pyplot as plt
 
## 데이터
age_category = month['월'] ## x축 눈금에 표시될 라벨
num_patient = month['판매금액']
mean_sales=round(month['판매금액'].mean(),2)
fig = plt.figure(figsize=(10,10)) ## 캔버스 생성
ax = fig.add_subplot() ## 그림 뼈대(프레임) 생성

## 시각화

colors = ['dodgerblue','lightcoral','lightcoral','dodgerblue','lightcoral','dodgerblue','dodgerblue','lightcoral','dodgerblue','lightcoral','dodgerblue','dodgerblue'] ## 색상 지정
xtick_label_position = list(range(len(age_category))) ## x축 눈금 라벨이 표시될 x좌표
plt.yticks(fontsize=18)
plt.xticks(xtick_label_position, age_category,fontsize=18) ## x축 눈금 라벨 출력

ax.axhline(mean_sales,label='Mean',color='black') ## 평균값을 y좌표로 하는 수평선 생성
 
ylim = ax.get_ylim() ## 기존의 y축 범위를 저장
 
## 평균값을 y 눈금에 추가한다.
yticks = list(ax.get_yticks())
yticks.append(mean_sales)
yticks = sorted(yticks)
ax.text(-0.9,mean_sales+0.1,f'Mean:{mean_sales}',fontsize=13)
 
bar=plt.bar(xtick_label_position, num_patient, color=colors) ## 바 차트 출력
 
plt.title('월별 판매 금액',fontsize=25) ## 타이틀 출력
plt.xlabel('월',fontsize=18) ## x축 라벨 출력
plt.ylabel('판매금액 (단위:억)',fontsize=18) ## y축 라벨 출력

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 15)
plt.savefig('./fig1.jpg',dpi=300)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 1. 기본 스타일 설정
plt.rcParams['font.size'] = 12

# 2. 데이터 준비
x = year['년']
y1 = year['판매금액']
y2 = year['상승량']

# 3. 그래프 그리기
fig, ax1 = plt.subplots()


ticklabel=['2020','2021','2022']
plt.xticks(x,ticklabel)
ax1.set_ylabel('판매금액 (단위:억)')
bar=ax1.bar(x, y1, color='crimson', alpha=0.7, width=0.7)
ax1.set_ylim(0, 70)
ax1.set_xlabel('연도')

ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.plot(x, y2, '-s', color='black', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax2.set_ylim(0, 70)
ax2.tick_params(axis='y', direction='in')


plt.title('연도별 판매 금액', fontsize=15)

for rect in bar:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 15)

ax.legend(loc='upper right', frameon=False, fontsize = 15)


plt.savefig('./fig2.jpg',dpi=300)
plt.show()
```
