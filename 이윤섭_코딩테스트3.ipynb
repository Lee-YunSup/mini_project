## 문제 4

1. 데이터 이산화가 가능한 column을 자동으로 찾아주는 함수를 만든다
    
    해당 함수의 검증은 **data_clinical_patient.txt** 파일로 한다

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder=LabelEncoder() 
```

```python
df1=pd.read_csv('./data_clinical_patient.txt',sep='\t',comment='#', encoding='cp949')
```

```python
# 라벨 인코딩을 실시했을 때 0,1로 분류 가능한 조건을 걸어 이산화 가능한 컬럼을 찾는다.
def find_column(x):
    for i in x.columns:
        items=x[i].unique()
        encoder.fit(items)
        if len(encoder.classes_)==2:
            print('데이터 이산화가 가능한 column:',i)
```

```python
find_column(df1)
```

```python
pd.set_option('display.max_columns', 500) # 보이는 컬럼 갯수를 500개로 조절
table_rna=pd.merge(df1,df2, left_on='PATIENT_ID',right_on='Hugo_Symbol')
table_rna
```

```python
NA=table_rna[table_rna['TUMOR_STATUS'].str.contains('Not')].index
table_rna=table_rna.drop(NA) #TUMOR_STATUS에 not이 들어간 변수 제거
```

```python
from statannot import add_stat_annotation
from catboost import CatBoostClassifier, Pool 
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
encoder=LabelEncoder() 
```

```python
table_rna=table_rna.T.reset_index().drop_duplicates(['index']).T # 중복된 열 제거
table_rna=table_rna.drop(index=0) # 0번째 인덱스행 제거
table_rna_columns = table_rna.iloc[0] # 첫번째 열음 컬럼명으로 사용하기 위해 값 저장
table_rna.columns=table_rna_columns # 저장한 값을 컬럼명으로 사용
```

```python
table_rna=table_rna.drop(index='index') # 0번째 인덱스행 제거
```

```python
table_rna=table_rna.reset_index() # 인덱스 번호 초기화
table_rna
```

```python
items=table_rna['TUMOR_STATUS'].unique() 
encoder.fit(items) # risk 변수 값들을 기준으로 라벨인코딩 실시한다.
```

```python
labels=encoder.transform(items)
```

```python
encoder.inverse_transform([0,1]) # 숫자를 문자로 인풋을 리스트로 해줌
```

```python
table_rna['TUMOR_STATUS'] = table_rna['TUMOR_STATUS'].apply(lambda x: encoder.transform([x])[0]) # risk에 라벨 링코딩을 설정해줌 
```
